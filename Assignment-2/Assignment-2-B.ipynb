```python
#A2-B-Q1

import pandas as pd
from sklearn.model_selection import train_test_split

iris_data = pd.read_csv('iris.csv')

print("Original dataset:")
print(iris_data.head())

X = iris_data.drop('variety', axis=1)
y = iris_data['variety']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

print("\nShapes of training and test sets:")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#A2-B-Q2

import pandas as pd
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('data_preprocess.csv')

print("Original dataset:")
print(data.head())

all_col = data.select_dtypes(include=['number']).columns

scaler = StandardScaler()

data[all_col] = scaler.fit_transform(data[all_col])

print("\nScaled dataset:")
print(data.head())
```

```python
#A2-B-Q1

import pandas as pd
from sklearn.model_selection import train_test_split

iris_data = pd.read_csv('Hospital_patient.csv')

print("Original dataset:")
print(iris_data.head())

X = iris_data.drop('Fever', axis=1)
y = iris_data['Fever']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print("\nShapes of training and test sets:")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#A2-B-Q2

import pandas as pd
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('Hospital_patient.csv')

print("Original dataset:")
print(data.head())

X = data.drop('Name', axis=1)

numeric_columns = X.select_dtypes(include=['float64', 'int64']).columns

scaler = StandardScaler()

X[numeric_columns] = scaler.fit_transform(X[numeric_columns])

print("\nScaled dataset:")
print(X.head())
```
